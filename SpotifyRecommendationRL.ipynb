```python
# Now use it with get_embeddings
import numpy as np
# For artist embeddings (handling collaborations)
def get_artist_embedding(artist_list):
    # Get embeddings for each artist in the list
    artist_embeds = [artist_encoded.get(artist) for artist in artist_list if artist in artist_encoded]

    # If no embeddings were found, return None or a default
    if not artist_embeds:
        return None

    # Average the embeddings
    return np.mean(artist_embeds, axis=0)

# Apply the function to get combined embeddings for artists
df['artist_embedding'] = df['artists'].apply(get_artist_embedding)# For album embeddings
```

```python
combined_vectors = []
num_cols = ['danceability', 'energy', 'key_1', 'key_2', 'key_3', 'key_4', 'key_5', 'key_6', 'key_7', 'key_8', 'key_9', 'key_10', 'key_11', 'mode', 'speechiness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']
embedding_cols = ['artist_embedding', 'album_embedding', 'track_embedding', 'genre_embedding']

# Loop through each row of the dataframe
for i in range(len(df)):
    # Collect numerical values from the row (converted to float)
    num_values = df.iloc[i][num_cols].values.astype(float)

    # Collect embedding vectors from the row
    emb_vectors = []
    for col in embedding_cols:
        emb_vectors.append(df.iloc[i][col])  # Append embedding for the current row

    # Combine numerical features and embedding vectors
    combined_features = np.concatenate([num_values] + emb_vectors)
    combined_vectors.append(combined_features)

# Stack the combined vectors into a matrix
combined_matrix = np.vstack(combined_vectors)
```

```python
from scipy.spatial.distance import pdist, squareform
from annoy import AnnoyIndex
import numpy as np
embedding_dim=combined_matrix.shape[1]
previous_track=[]
current_track=[]
annoy_index = AnnoyIndex(embedding_dim, 'angular')
for i in range(len(df)):
    annoy_index.add_item(i, combined_matrix[i])
annoy_index.build(10)
def get_recommended_tracks(current_track_index, k=5):
    print("Fetchingg")
    nearest_neighbors = annoy_index.get_nns_by_item(current_track_index, k)
    neighbor_embeddings = np.array([combined_matrix[i] for i in nearest_neighbors])
    pairwise_distances = pdist(neighbor_embeddings, metric='cosine')  # Using cosine distance (angular)

    # Diversity score as the average of the pairwise distances
    diversity_score = pairwise_distances.mean()

    print(f"Diversity Score for Track {current_track_index}: {diversity_score}")

    previous_track.extend(nearest_neighbors)
    current_track=nearest_neighbors[0]
    return nearest_neighbors

```

RL BASED RECOMMENDER

```python
state = {
    'previous_track': previous_track,
    'current_track': current_track
}
```

```python
import random
from annoy import AnnoyIndex

class MusicRecommenderEnv:
    def __init__(self, annoy_index, combined_matrix, epsilon=0.1):
        self.annoy_index = annoy_index
        self.combined_matrix = combined_matrix
        self.epsilon = epsilon
        self.previous_tracks = []
        self.current_track = None
        self.max_distance = self.calculate_max_distance()
        self.threshold = self.max_distance * 0.3


    def calculate_max_distance(self):
          max_distance = -1
          for i in range(len(self.combined_matrix)):
              for j in range(i + 1, len(self.combined_matrix)):
                  distance = self.annoy_index.get_distance(i, j)
                  max_distance = max(max_distance, distance)
          return max_distance
    def reset(self):
        self.previous_tracks = []
        self.current_track = random.choice(range(len(self.combined_matrix)))
        return {
            'previous_tracks': self.previous_tracks,
            'current_track': self.current_track
        }
    def step(self, action, calculate_reward):
        reward = calculate_reward(self.previous_tracks, self.current_track, action, self.threshold, self.annoy_index)

        self.previous_tracks.append(self.current_track)
        self.current_track = action

        return {
            'previous_tracks': self.previous_tracks,
            'current_track': self.current_track
        }, reward


    def start_action(self, state, epsilon=None):
      if epsilon is None:
          epsilon = self.epsilon

      recommended_tracks = state['previous_tracks']

      # If no previous tracks, randomly select from full dataset
      if not recommended_tracks:
          return random.choice(range(len(self.combined_matrix)))

      if random.random() < epsilon:
          action = random.choice(recommended_tracks)
      else:
          action = recommended_tracks[0]

      return action



```

```python
def calculate_reward(previous_tracks, current_track, action, threshold, annoy_index):
    if not previous_tracks:
        return 1

    distances = [annoy_index.get_distance(prev, action) for prev in previous_tracks]
    avg_distance = sum(distances) / len(distances)

    lower_bound = threshold * 0.5
    upper_bound = threshold * 1.2

    if lower_bound <= avg_distance <= upper_bound:
        return 1
    else:
        return 0


rl_recommended_tracks=[]

env = MusicRecommenderEnv(annoy_index, combined_matrix, epsilon=0.1)

state = env.reset()

for episode in range(5):
    print(f"Episode {episode + 1}:")

    action = env.start_action(state)
    print(f"Recommended track: {action}")

    rl_recommended_tracks.append(action)

    next_state, reward = env.step(action, calculate_reward)
    print(f"Reward: {reward}")

    state = next_state
```

```python
def diversity_score(previous_tracks, current_track, annoy_index):
    if not previous_tracks:
        return 0.0  # No previous tracks to compare with

    distances = [
        annoy_index.get_distance(prev, current_track) for prev in previous_tracks
    ]
    return sum(distances) / len(distances)
```

```python
rl_div = diversity_score(state['previous_tracks'], action, annoy_index)
print(f"Diversity Score for current recommended track: {rl_div:.4f}")
```

```python
import seaborn as sns

def plot_distance_heatmap(recommended_indices):
    # Extract embeddings for recommended tracks
    embeddings = np.array([combined_matrix[i] for i in recommended_indices])
    pairwise_cosine = squareform(pdist(embeddings, metric='cosine'))

    # Plot heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(pairwise_cosine, annot=True, cmap='viridis', xticklabels=recommended_indices, yticklabels=recommended_indices)
    plt.title("Cosine Distance Heatmap of Recommended Tracks")
    plt.xlabel("Track Index")
    plt.ylabel("Track Index")
    plt.show()
```
